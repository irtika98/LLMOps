## Evaluate the retrieval

Defining the expected ranking and evaluating the responses we actually get

```python
import scipy.stats as stats

# For each query, you can define what the expected ranking is.
EXPECTED_RANKING = {
    "Which countinent does india belong?": ["id1", "id3", "id2"],
    "What is the weather today": ["id2", "id1", "id3"],
    "Do you know spanish?": ["id3", "id1", "id2"],
    "What is the capital of Spain?": ["id3", "id1", "id2"]
}


def measure_correlation(row: "pandas.core.series.Series", ranking_column_name: str = "top doc ids") -> float:
    r"""
    A simple test that compares the expected ranking for a given query with the actual ranking produced
    by the embedding function being tested.
    """
    input_query = row["query_texts"]
    correlation, _ = stats.spearmanr(row[ranking_column_name], EXPECTED_RANKING[input_query])
    return correlation
```

```python
experiment.evaluate("ranking_correlation", measure_correlation)
```

```python
experiment.visualize()
```

# Retrieval Augmented Generation Experiment Example

```python
from prompttools.experiment import OpenAIChatExperiment
from prompttools.harness import RetrievalAugmentedGenerationExperimentationHarness
```

```python
def _extract_doc_from_row(row: 'pandas.core.series.Series'):
    return row['documents']


def _extract_query_metadata_from_row(row: 'pandas.core.series.Series'):
    return f"emb_fn: {row['embed_fn']}, prompt: {row['query_texts']}"
```

```python
chat_models = ["gpt-3.5-turbo"]
messages = [
    [
        {"role": "system", "content": "You are a helpful assistant.You answer only based on the documents you get in prompt. If the Prompt doesn't contain relevant information for answering the question, you say 'Retrieved documents dont have relevant information for answering the query'"},
        {"role": "user", "content": "Which countinent does india belong?"},
    ],
    [
        {"role": "system", "content": "You are a helpful assistant.You answer only based on the documents you get in prompt. If the Prompt doesn't contain relevant information for answering the question, you say 'Retrieved documents dont have relevant information for answering the query'"},
        {"role": "user", "content": "What is the capital of Spain?"},
    ]
]
temperatures = [1.0]  # You can test multiple temperate or other parameters as wel
# You can add more parameters that you'd like to test here.

llm_chat_arguments = {"model": chat_models, "messages": messages, "temperature": temperatures}
```

```python
query_collection_params = {"query_texts": ["Which countinent does india belong?", "What is the weather today", "Do you know spanish?", "What is the capital of Spain?"], "n_results": [1]}

experiment = ChromaDBExperiment(
    chroma_client,
    collection_name,
    use_existing_collection,
    query_collection_params,
    emb_fns,
    emb_fns_names,
    add_to_collection_params,
)
```

```python
rag_experiment = RetrievalAugmentedGenerationExperimentationHarness(
    vector_db_experiment = experiment,
    llm_experiment_cls = OpenAIChatExperiment,
    llm_arguments = llm_chat_arguments,
    extract_document_fn = _extract_doc_from_row,
    extract_query_metadata_fn = _extract_query_metadata_from_row,
)

rag_experiment.run()
rag_experiment.visualize()
```

Auto-evaluation can be done also for RAG experiments using  ( I didn't run this snippet because it uses GPT-4 )

```python
from prompttools.utils import autoeval_with_documents

documents = ["India is a country in Asia",
             "Today it will rain",
             "Madrid is the capital of Spain."]

rag_experiment.evaluate("Score", autoeval_with_documents, documents=[documents] * 16)
rag_experiment.visualize()
```

# Lang Chain Sequential Chain Experiment Example

```python
!pip install langchain
```

```python
from prompttools.experiment import SequentialChainExperiment
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate

from prompttools.utils import semantic_similarity


llm = ChatOpenAI(temperature=0.9)

# prompt template 1: translate to english
first_prompt = ChatPromptTemplate.from_template("Translate the following review to english:" "\n\n{Review}")

second_prompt = ChatPromptTemplate.from_template(
    "Can you summarize the following review in 1 sentence:" "\n\n{English_Review}"
)

third_prompt = ChatPromptTemplate.from_template("What language is the following review:\n\n{Review}")

fourth_prompt = ChatPromptTemplate.from_template(
    "Write a follow up response to the following "
    "summary in the specified language:"
    "\n\nSummary: {summary}\n\nLanguage: {language}"
)

experiment = SequentialChainExperiment(
    llm=[ChatOpenAI],
    prompt=["मुझे शावरमा खाना बहुत पसंद है", "me encanta comer shawarma"],
    prompt_template=[[first_prompt, second_prompt, third_prompt, fourth_prompt]],
    **{
        "temperature": [0.1, 0.9],
        "input_variables": [["Review"]],
        "output_variables": [["English_Review", "summary", "followup_message"]],
        "output_key": [["English_Review", "summary", "language", "followup_message"]],
    },
)


experiment.run()

experiment.evaluate("similar_to_expected", semantic_similarity, expected=["I love to eat shawarma"] * 4)

experiment.visualize()
```
